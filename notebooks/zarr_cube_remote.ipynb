```python
from xcube import open_from_obs
import xarray as xr
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
ds = open_from_obs('dcs4cop-obs-01/OLCI-SNS-RAW-CUBE-2.zarr', endpoint_url='http://obs.eu-de.otc.t-systems.com')
```

```python
ds
```

```python
ds.conc_chl.isel(time=62).plot.imshow()
```

```python
ds.conc_chl.sel(time='2017-09-16', method='nearest')
```

```python
ds.conc_chl.sel(lon=1., lat=50.5, method='nearest').plot.line('r-o')
```

```python
chl_w = ds.conc_chl.resample(time='1W').mean(dim='time')
```

```python
chl_w.isel(time=20).plot.imshow()
```

```python
chl_w_ts = chl_w.sel(lon=1., lat=50.5, method='nearest')
```

```python
chl_w_ts.load()
```

```python
chl_w_ts.plot.line('r-o')
```

```python
chl_w_ts_clean = chl_w_ts.where(~np.isnan(chl_w_ts), drop=True)
```

```python
chl_w_ts_clean
```

```python
chl_w_ts_clean.plot.line('r-o')
```

```python
chl_w_ts_clean.sel(time='2018-08-16', method='nearest')
```
